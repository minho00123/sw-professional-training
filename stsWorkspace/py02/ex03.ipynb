```python
class Cl01:
    def __init__(self,name='user01',age=25):             #생성자
        self.name=name
        self.age=age
        self.__email='test@test'
    def func01(self):
        print(self,'메서드',self.name,self.age,self.__email)        #메서드
    def func02(self,su):
        print(self,'메서드:',su)
```

```python
obj=Cl01()
obj.func01()
obj.func02(1234)
print(obj.name,obj.age)
```

```python
class Cl03:
    arr1=[]
    @classmethod
    def func01(cls):
        print(cls.arr1)
```

```python
cl3=Cl03()
cl3.func01()
```

```python
class Cl04:
    def __init__(self,su):
        self.su1=su
    def func1(self):
        print('cl04 func01')
class Cl05:
    def func1(self):
        print('cl05 func01')
    def func2(self):
        print('cl05 func02')

class Cl06(Cl04,Cl05):
    def __init__(self,su):
        super().__init__(su)
        # pass

cl6=Cl06(3333)
cl6.func1()
cl6.func2()
print(cl6.su1)
```
